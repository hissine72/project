```python
import numpy as np
import pandas as pd

df1 = pd.read_csv(r"C:\Users\ahmed\Videos\Project\webscraping\cars1.csv", encoding='ISO-8859-1')
df1.head()
```

```python
df = df1.copy()
```

```python
#Replacing None with null values
df.replace(to_replace=['None'], value=np.nan, inplace=True)
```

```python
df.isnull().sum()
```

```python
#Dropping rows with null values in "price" column
df.dropna(subset=["Price"], axis=0, inplace=True)
df.isnull().sum()
```

```python
#drop null year
df.dropna()
```

```python
print(df.dtypes)
```

```python
#Fixing number formats
df['Price'] = df['Price'].str.replace(',', '', regex=True)
```

```python
#Replacing any text with 0
#Changing column data type to float
df['Price'] = df['Price'].str.replace('\W', '0', regex=True).str.replace('[A-Za-z]', '0', regex=True).astype(float)
```

```python
#Statistical summary of the column
df['Price'].describe()
```

```python
#Repeating the proccess for KM column
df['Km'] = df['Km'].str.replace(',', '', regex=True)
```

```python
df['Km'] = df['Km'].str.replace('\W', '0', regex=True).str.replace('[A-Za-z]', '0', regex=True).astype(float)
```

```python
df['Km'].describe()
```

```python
df['Year'] = df['Year'].astype(float)
```

```python
df['Year'].describe()
```

```python
df[df==0].count()
```

```python
#Drop any empty price value
df= df[df['Price'] != 0]
df[df==0].count()
```

```python
#Replace unknown KM values with the mean
avg_km = df["Km"].mean(axis=0)
df["Km"].replace(0, avg_km, inplace=True)
df[df==0].count()
```

```python
#selecting the makes with low number of data
a = list(map(list, df["Make"].value_counts().items()))
d = [a[m][0] for m in range(len(a)) if a[m][1] < 100 ]
d
```

```python
#Dropping them
df = df[~df['Make'].isin(d)]
```

```python
df.Transmission.value_counts()
```

```python
#Dropping Transmssion types with low number of data
df = df[(df['Transmission'] != 'CVT') & (df['Transmission'] != 'DSG')]
```

```python
df.Fuel.value_counts()
```

```python
#Dropping Fuel types with low number of data
df = df[(df['Fuel'] != 'Hybrid') & (df['Fuel']!='Electric')]
```

```python
# Number of uniques values in Data
pd.DataFrame({'': df.columns , 'Number of unique values':  df.nunique().values})
```

```python
#extracting all models to a list
b = list(map(list, df["Model"].value_counts().items()))
b
```

```python
#selecting the models with low number of data
b = list(map(list, df["Model"].value_counts().items()))
f = [b[m][0] for m in range(len(b)) if b[m][1] < 9 ]
f
```

```python
#Dropping them
df = df[~df['Model'].isin(f)]
```

```python
#selecting the years with low number of data
p = list(map(list, df["Year"].value_counts().items()))
g = [p[m][0] for m in range(len(p)) if p[m][1] < 9 ]
g
```

```python
#Dropping them
df = df[~df['Year'].isin(g)]
```

```python
df.Year.value_counts()
```

```python
#Creating lists of years
the70s = [n for n in range(1970,1980)]
the80s = [n for n in range(1980,1990)]
the90s = [n for n in range(1990,2000)]
print(the90s)
```

```python
#Gathering them into groups
df.loc[df['Year'].isin(the70s) , 'Year'] = 70
df.loc[df['Year'].isin(the80s) , 'Year'] = 80
df.loc[df['Year'].isin(the90s) , 'Year'] = 90
```
